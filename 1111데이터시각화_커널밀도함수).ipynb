```python
import pandas as pd
import matplotlib as mpl
import numpy as np
mpl.rcParams['axes.unicode_minus'] = False


from pandas.plotting import scatter_matrix
```

```python
df = pd.DataFrame(np.random.randn(1000, 4), columns=['A','B','C','D'])
```

```python
df.head()
```

# Density Plot
* df.plot.kde()로 그릴 수 있음
* 히스토그램과 기능은 비슷한데, 곡선으로 보여준다는 점이 다르다
* 히스토그램은 y축이 빈도수, density plot은 그래프의 넓이를 1로 변환했을 때 + 정규분포화 하는 것
* 히스토그램의 이산적인 형태를 연속적인 (밀도) 형태로 나타내는 것

```python
# alpha는 투명도를 의미함
# kde는 커널 밀도 함수 = 대각선은 같은 컬럼의 교차지점이니까 여기를 커널밀도함수로 채우라는 것

g = scatter_matrix(df, alpha=0.2, figsize = (6,6), diagonal ='kde')
```

```python
# density와 kde는 둘다 동일함

df['A'].plot.kde()
```

```python
df['A'].plot.hist()
```
